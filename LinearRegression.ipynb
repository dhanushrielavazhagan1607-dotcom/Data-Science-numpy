```python
import pandas as pd
import numpy as np
```

```python
d=pd.read_csv("Salary_Data.csv")
```

```python
d
```

```python
d.columns[d.isna().any()]
```

```python
d[["Years of Experience","Salary"]]=d[["Years of Experience","Salary"]].fillna(d[["Years of Experience","Salary"]].mean())
```

```python
d=d[["Years of Experience","Salary"]]
```

```python
pip install Scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=d[["Years of Experience"]]
y=d["Salary"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pre=lr.predict(x_test)
y_pre
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pre)
```

```python
lr.coef_
```
